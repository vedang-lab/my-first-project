```python
#write a function named as max_min to find minimum (or maximum) element in an array
```

```python
def max_min(arr, find="max"):

    if not arr:
        return None

    if find == "max":
        result = arr[0]
        for num in arr:
            if num > result:
                result = num
        return result

    elif find == "min":
        result = arr[0]
        for num in arr:
            if num < result:
                result = num
        return result

    else:
        raise ValueError("Parameter 'find' must be either 'max' or 'min'.")

arr = [12, 3, 4, 15, 45, 78, 450]
print("Maximum:", max_min(arr, "max"))
print("Minimum:", max_min(arr, "min"))
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
```

```python
def Counting_EvenOdd(arr):

    count_even = 0
    count_odd = 0

    for num in arr:
        if num % 2 == 0:
            count_even += 1
        else:
            count_odd += 1

    return count_even, count_odd

arr = [12, 1234, 45, 67, 1, 90, 33]
even_count, odd_count = Counting_EvenOdd(arr)
print("Even numbers:", even_count)
print("Odd numbers:", odd_count)
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
```

```python
def Binary_Search(arr, target):

    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1
```
